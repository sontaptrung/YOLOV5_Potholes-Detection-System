```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
data = pd.read_csv('fps_.txt', header=None)
data.rename(columns={0: 'fps'}, inplace=True)
```

```python
data = data[0:300]
data.shape
```

```python
sns.set_style('darkgrid')
sns.palettes.color_palette('muted')
sns.lineplot(data=data, x=data.index, y='fps',color=sns.palettes.color_palette('muted')[0])
plt.title('FPS')
plt.legend(['FPS'])
```
